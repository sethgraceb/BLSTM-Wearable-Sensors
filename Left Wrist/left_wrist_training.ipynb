# Left wrist training

## Acceleration

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('acc_data (left_wrist).csv')

X = np.column_stack((np.array(df['Acc_X']), np.array(df['Acc_Y']), np.array(df['Acc_Z'])))
y = np.array(df['PacketCounter']).reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)

#print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state = 42)
#print(X)
```

```python
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y.ravel())
```

```python
ypred = regressor.predict(X)
print(ypred)
```

### Acc_X

```python
df_new = pd.DataFrame()
df_new['Acc_X'] = df['Acc_X']

fig = plt.figure(figsize=(10, 5))
sns.lineplot(data=df_new)
plt.title('Acceleration X', fontsize = 25)
plt.legend(prop={'size': 10})
plt.xlabel('', fontsize = 25)
plt.tick_params(labelsize=25)
plt.show()
```

### Acc_Y

```python
df_new = pd.DataFrame()
df_new['Acc_Y'] = df['Acc_Y']

fig = plt.figure(figsize=(10, 5))
sns.lineplot(data=df_new)
plt.title('Acceleration Y', fontsize = 25)
plt.legend(prop={'size': 10})
plt.xlabel('', fontsize = 25)
plt.tick_params(labelsize=25)
plt.show()
```

### Acc_Z

```python
df_new = pd.DataFrame()
df_new['Acc_Z'] = df['Acc_Z']

fig = plt.figure(figsize=(10, 5))
sns.lineplot(data=df_new)
plt.title('Acceleration Z', fontsize = 25)
plt.legend(prop={'size': 10})
plt.xlabel('', fontsize = 25)
plt.tick_params(labelsize=25)
plt.show()
```

```python
df_new = pd.DataFrame()
df_new['Acc_X'] = df['Acc_X']
df_new['Acc_Y'] = df['Acc_Y']
df_new['Acc_Z'] = df['Acc_Z']

fig = plt.figure(figsize=(10, 5))
sns.lineplot(data=df_new)
plt.title('Acceleration', fontsize = 25)
plt.legend(prop={'size': 10})
plt.xlabel('', fontsize = 25)
plt.tick_params(labelsize=25)
plt.show()
```

```python
df_new = pd.DataFrame()
df_new['Acc_X'] = df['Acc_X']
df_new['Acc_Y'] = df['Acc_Y']
df_new['Acc_Z'] = df['Acc_Z']
df_new['Predicted'] = ypred

fig = plt.figure(figsize=(10, 5))
sns.lineplot(data=df_new)
plt.title('Acceleration', fontsize = 25)
plt.legend(prop={'size': 10})
plt.xlabel('', fontsize = 25)
plt.tick_params(labelsize=25)
plt.show()
```

#### Baseline

```python
#Baseline
from sklearn.dummy import DummyRegressor
dummy = DummyRegressor(strategy="mean")
dummy.fit(X, y)
y_dummypred = dummy.predict(X)
```

```python
df_new = pd.DataFrame()
df_new['Acc_X'] = df['Acc_X']
df_new['Acc_Y'] = df['Acc_Y']
df_new['Acc_Z'] = df['Acc_Z']
df_new['Baseline'] = y_dummypred
df_new['Predicted'] = ypred


fig = plt.figure(figsize=(10, 5))
sns.lineplot(data=df_new)
plt.title('Acceleration', fontsize = 25)
plt.legend(prop={'size': 15})
plt.xlabel('', fontsize = 25)
plt.tick_params(labelsize=25)
plt.show()
```
