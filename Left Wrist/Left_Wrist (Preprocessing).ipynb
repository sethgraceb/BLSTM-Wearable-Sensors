# Acceleration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

dataf = pd.read_csv("Left_Wrist.csv", error_bad_lines=False)
print(dataf.columns)
dataf = dataf.drop(columns=['SampleTimeFine', 'Euler_X', 'Euler_Y', 'Euler_Z', 'Gyr_X', 'Gyr_Y', 'Gyr_Z'])
```

```python
print(dataf.columns)
```

```python

dataf_total = dataf.copy()
dataf_total = dataf_total.dropna()

dataf_new = pd.DataFrame()
dataf_new['Acc_X'] = dataf_total['Acc_X'].diff()
dataf_new['Acc_Y'] = dataf_total['Acc_Y'].diff()
dataf_new['Acc_Z'] = dataf_total['Acc_Z'].diff()
```

```python
fig = plt.figure(figsize=(10,5))
sns.lineplot(data=dataf_total)
plt.legend(prop={'size': 15})
plt.title('Acceleration', fontsize = 25)
plt.xlabel('', fontsize=15)
plt.ylabel('', fontsize=15)
plt.tick_params(labelsize=25)
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
sns.lineplot(data=dataf_new)
plt.legend(prop={'size': 15})
plt.title('Acceleration', fontsize = 25)
plt.xlabel('', fontsize=30)
plt.ylabel('', fontsize=30)
plt.tick_params(labelsize=25)
plt.show()
```

```python
dataf.to_csv('acc_data (left_wrist).csv', index=False)
```
